```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
```

```python
def process_mat(path, ch=4):
    data = sio.loadmat(path)
    length = len(data['data'][0])
    data_array = np.array(data['data'][0])
    data_array = data_array.reshape([length/ch, ch])
    return data_array
```

```python
data_array = process_mat('step_debug_1630346885.13_reward.mat', ch=3)
```

```python
plt.plot(data_array[:, 2], data_array[:, 1])
plt.title('Time vs Reward for Original DDPG')
```

```python
plt.plot(data_array[:, 0], data_array[:, 1])
plt.title('Step Number vs Reward for Original DDPG')
```
